```python
#
#  Create spark view
# 

# run 
make_view = False

# drop existing view
stmt_del = """
  DROP VIEW lh_snp_500.gold_stock_data;
"""

if make_view:
  try:
    spark.sql(stmt_del)
  except:
    pass


# create new view
stmt_add = """
  CREATE VIEW gold_stock_data AS
    SELECT * FROM silver_stocks_2013_full 
    UNION
    SELECT * FROM silver_stocks_2014_full 
    UNION
    SELECT * FROM silver_stocks_2015_full 
    UNION
    SELECT * FROM silver_stocks_2016_full 
    UNION
    SELECT * FROM silver_stocks_2017_full 
"""

if make_view:
  try:
    spark.sql(stmt_add)
  except:
    pass
```

```python
%%sql

--
-- Sample view
--

select * from gold_stock_data limit 5
```
