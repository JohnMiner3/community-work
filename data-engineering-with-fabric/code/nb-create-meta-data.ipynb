```python
%%sql

--
--  remove table
--

drop table if exists meta_data;
```

```python
%%sql

--
--  create table
--

create table meta_data 
(
    pipeline_id int,
    container_nm string,
    file_nm string,
    header_flg boolean,
    delimiter_str string,
    table_nm string,
    schema_str string
);
```
