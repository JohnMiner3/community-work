```python
#
#  Remove all data tables, skip meta data
#

# get list (works with schema less lakehouses)
# tables = spark.sql('show tables;').collect()

# work around
tables = [f.name for f in mssparkutils.fs.ls("Tables/bronze")]

# for each table name
for table in tables:

    # skip meta data table
    if table == 'meta_data':
        print('do not drop the meta data table')
    else:
        try:
            stmt = "drop table bronze." + table
            print(stmt)
            ret = spark.sql(stmt)
            
        except:

            try:
                stmt = "drop view bronze." + table
                print(stmt)
                ret = spark.sql(stmt)
            except:
                pass
            
```
