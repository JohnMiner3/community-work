```python
#
#  Show data files
#

mssparkutils.fs.ls("Files/weather")
```

```python
#
#  Create high temp view
#

df1 = spark.read.format("csv").option("header","true").load("Files/weather/high_temps.csv")
df1.createOrReplaceTempView("vw_high_temps")
```

```python
#
#  Create low temp view
#

df2 = spark.read.format("csv").option("header","true").load("Files/weather/low_temps.csv")
df2.createOrReplaceTempView("vw_low_temps")
```

```python
%%sql
select 
  l.date as obs_date,
  l.temp as obs_low_temp,
  h.temp as obs_high_temp
from vw_low_temps as l 
join vw_high_temps as h 
on l.date = h.date
limit 10
```

```python
#
#  create bronze schema
#

stmt = "create schema if not exists bronze"
spark.sql(stmt)
```

```python
#
#  Drop old table
#

stmt = "drop table if exists bronze.spark_weather_data"
spark.sql(stmt)
```

```python
#
#  Create new table
#

stmt = """
select 
  l.date as obs_date,
  l.temp as obs_low_temp,
  h.temp as obs_high_temp
from vw_low_temps as l 
join vw_high_temps as h 
on l.date = h.date
"""

df3 = spark.sql(stmt)
df3.write.saveAsTable("bronze.spark_weather_data")
```
