```python
#
#  Remove all data tables, skip meta data
#

# get list
tables = spark.sql('show tables;').collect()

# for each table name
for table in tables:

    # skip meta data table
    if table.tableName == 'meta_data':
        print('do not drop the meta data table')
    else:
        try:
            stmt = "drop table " + table.tableName
            print(stmt)
            ret = spark.sql(stmt)
        except:
            stmt = "drop view " + table.tableName
            print(stmt)
            ret = spark.sql(stmt)
```
