```python
%%sql

 --
 --  1 - Create managed TEXT table
 --

-- del
drop table if exists txt_nasa_ex1;

-- add
 create table if not exists txt_nasa_ex1 as
   select * from text.`Files/Nasa/nasa_http_logs.txt`;
```

```python
%%sql

--
-- 2 - get table info
--

describe table extended txt_nasa_ex1
```

```python
%%sql

 --
 --  3 - Create unmanaged TEXT table
 --

-- del
drop table if exists txt_nasa_ex2;

-- add
create table if not exists txt_nasa_ex2
using text 
location "Files/Nasa/nasa_http_logs.txt"
```

```python
%%sql

--
-- 4 - get table info
--

describe table extended txt_nasa_ex2
```

```python
%%sql

--
-- 5 - show sample records
--

select * from txt_nasa_ex1 limit 10
```

```python
%%sql

--
-- 6 - find http errors
--

with cte as (
select 
    row_number() over (order by 1) as rn,
    replace(split_part(value, '|', 3), '"', '') as method, 
    split_part(value, '|', 5) as url,
    split_part(value, '|', 6) as response 
from txt_nasa_ex1
)
select * from cte where rn > 1 and response <> 200 limit 10
```
