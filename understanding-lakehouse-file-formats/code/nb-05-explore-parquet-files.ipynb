```python
%%sql

 --
 --  1 - Create managed PARQUET table
 --

-- delta columns can not have certain chars like spaces

-- del
drop table if exists parquet_stocks_ex1;

-- add
 create table if not exists parquet_stocks_ex1 as
     select 
        `Date` as _date,
        year(`Date`) as _year, 
        month(`Date`) as _month,
        Ticker as _ticker,
        Open as _open,
        High as _high,
        Low as _low,
        Close as _close,
        Volume as _volume,
        Dividends as _dividends,
        `Stock Splits` as _stock_splits
      from parquet.`Files/Stocks/all_stock_data.parquet`;
```

```python
%%sql

--
--  2 - show data
--

select * from parquet_stocks_ex1 limit 10
```

```python
%%sql

--
-- 3 - get table info
--

describe table extended parquet_stocks_ex1
```

```python
%%sql

 --
 --  4 - Create unmanaged PARQUET table
 --

-- del
drop table if exists parquet_stocks_ex2;

-- add
create table if not exists parquet_stocks_ex2
using parquet 
location "Files/Stocks/all_stock_data.parquet"
```

```python
%%sql

--
-- 5 - get table info
--

describe table extended parquet_stocks_ex2
```

```python
%%sql

--
-- 5 - show sample records
--

select * from parquet_stocks_ex2 limit 10
```

```python

%%sql

--
--  6 - counts should be the same regardless of format
--

select 'managed table using sql', count(*) as rec_cnt from parquet_stocks_ex1
union
select 'unmanaged table using sql', count(*) as rec_cnt from parquet_stocks_ex2
```
