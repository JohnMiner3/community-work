```python
%%sql

 --
 --  1 - Create managed AVRO table
 --

-- del
drop table if exists orc_stocks_ex1;

-- add
 create table if not exists orc_stocks_ex1 as
     select * from orc.`Files/Stocks/all_stock_data.orc`;
```

```python
%%sql

--
--  2 - show data
--

select * from orc_stocks_ex1 limit 10
```

```python
%%sql

--
-- 3 - get table info
--

describe table extended orc_stocks_ex1
```

```python
%%sql

 --
 --  4 - Create unmanaged AVRO table
 --

-- del
drop table if exists orc_stocks_ex2;

-- add
create table if not exists orc_stocks_ex2
using orc
location "Files/Stocks/all_stock_data.orc"
```

```python
%%sql

--
-- 5 - get table info
--

describe table extended orc_stocks_ex2
```

```python
%%sql

--
-- 6 - show sample records
--

select * from orc_stocks_ex2 limit 10
```

```python

%%sql

--
--  7 - counts should be the same regardless of format
--

select 'managed table using sql', count(*) as rec_cnt from avro_stocks_ex1
union
select 'unmanaged table using sql', count(*) as rec_cnt from avro_stocks_ex2
```
